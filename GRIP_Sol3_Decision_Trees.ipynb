**1. Importing the libraries and the dataset**

```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = datasets.load_iris()

# Forming the X dataframe
X = pd.DataFrame(iris.data, columns=iris.feature_names)

# Forming the y matrix
y = iris.target
```

```python
X.isnull().sum()
```

**2. Forming the decision tree model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
clf = DecisionTreeClassifier(random_state=0)
clf = clf.fit(X_train, y_train)
```

```python
y_train_pred = clf.predict(X_train)
y_test_pred = clf.predict(X_test)
```

**3. Evaluating the model**

```python
from sklearn.metrics import r2_score, mean_squared_error as mse
def evaluate_model(name):
  rmse_tr = np.sqrt(mse(y_train, y_train_pred))
  print(f"{name} : RMSE Train of = {rmse_tr:.4f}")

  rmse_te = np.sqrt(mse(y_test, y_test_pred))
  print(f"{name} : RMSE Test of = {rmse_te:.4f}")

  r2_sc_train = r2_score(y_train, y_train_pred)
  print(f"{name} : R2 Score Train = {r2_sc_train:.4f}")

  r2_sc_test = r2_score(y_test, y_test_pred)
  print(f"{name} : R2 Score Test = {r2_sc_test:.4f}")
```

```python
evaluate_model("Decision Tree Classifier")
```

**4. Plotting the tree**

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```
